```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import MaxAbsScaler
```

```python
MinmaxScale=MinMaxScaler()
StandarScale=StandardScaler()
RobustScale=RobustScaler()
NormalScale=Normalizer()
MaxScale=MaxAbsScaler()
```

### Train Scaling

```python
X_train.head()
```

```python
index_cols = []
categorical_cols=['PerformanceRating','MonthlyIncome_cal','MonthlyRate_cal',
                   'YearsAtCompany_cal', 'TotalWorkingYears_cal','YearsSinceLastPromotion_cal', 'NumCompaniesWorked_cal','Age_cal','PerformanceScore_cal']
numeric_cols = list(X_train.columns.difference(index_cols + categorical_cols)) 
```

```python
X_train_num=X_train[numeric_cols]
```

```python
X_train_num.shape,X_train.shape
```

```python
X_train_num.index[np.isinf(X_train_num).any(1)]
```

```python
MinMax_Train=MinmaxScale.fit_transform(X_train_num)
Std_Train=StandarScale.fit_transform(X_train_num)
Robust_Train=RobustScale.fit_transform(X_train_num)
Normal_Train=NormalScale.fit_transform(X_train_num)
MaxAbs_Train=MaxScale.fit_transform(X_train_num)
```

```python
MinMax_Train=pd.DataFrame(MinMax_Train,columns=numeric_cols,index=X_train_num.index)
Std_Train=pd.DataFrame(Std_Train,columns=numeric_cols,index=X_train_num.index)
Robust_Train=pd.DataFrame(Robust_Train,columns=numeric_cols,index=X_train_num.index)
Normal_Train=pd.DataFrame(Normal_Train,columns=numeric_cols,index=X_train_num.index)
MaxAbs_Train=pd.DataFrame(MaxAbs_Train,columns=numeric_cols,index=X_train_num.index)
```

```python
required_cols=index_cols + categorical_cols
required_cols
```

```python
MinMax_Train=pd.merge(X_train[required_cols], MinMax_Train, left_index=True, right_index=True)
Std_Train=pd.merge(X_train[required_cols], Std_Train, left_index=True, right_index=True)
Robust_Train=pd.merge(X_train[required_cols], Robust_Train, left_index=True, right_index=True)
Normal_Train=pd.merge(X_train[required_cols], Normal_Train, left_index=True, right_index=True)
MaxAbs_Train=pd.merge(X_train[required_cols], MaxAbs_Train, left_index=True, right_index=True)
```

```python
MinMax_Train.shape,Std_Train.shape,Robust_Train.shape,Normal_Train.shape,MaxAbs_Train.shape
```
